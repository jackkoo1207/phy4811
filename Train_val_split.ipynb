```python
import numpy as np
import matplotlib.pyplot as plt
import os
dir_axion = r'Model_III/axion'
dir_cdm = r'Model_III/cdm'
dir_no = r'Model_III/no_sub'
count = 0
train_size = 3000
val_size = 600
test_size =600
count=0
image_size=64
```

```python
train_data=np.zeros((3*train_size, 1, image_size, image_size))
train_label=np.ones((3*train_size, 1, image_size, image_size))
train_label[:train_size, 0, :, :] *= 0
train_label[train_size:2*train_size, 0, :, :] *= 1
train_label[2*train_size:, 0, :, :] *= 2
for path in os.listdir(dir_axion):
    if count==train_size:
        break
    if os.path.isfile(os.path.join(dir_axion, path)):
        train_data[count, 0, :, :] = np.load(os.path.join(dir_axion, path), allow_pickle=True)[0]
        count += 1
for path in os.listdir(dir_cdm):
    if count==train_size*2:
        break
    if os.path.isfile(os.path.join(dir_cdm, path)):
        train_data[count, 0, :, :] = np.load(os.path.join(dir_cdm, path), allow_pickle=True)
        count += 1
for path in os.listdir(dir_no):
    if count==train_size*3:
        break
    if os.path.isfile(os.path.join(dir_no, path)):
        train_data[count, 0, :, :] = np.load(os.path.join(dir_no, path), allow_pickle=True)
        count += 1
print(train_data.shape)
```

```python
dir_axion = r'Model_III_test/axion'
dir_cdm = r'Model_III_test/cdm'
dir_no = r'Model_III_test/no_sub'
count=0
image_size=64
train_data=np.zeros((3*val_size, 1, image_size, image_size))
train_label=np.ones((3*val_size, 1, image_size, image_size))
train_label[:train_size, 0, :, :] *= 0
train_label[train_size:2*train_size, 0, :, :] *= 1
train_label[2*train_size:, 0, :, :] *= 2
for path in os.listdir(dir_axion):
    if count==val_size:
        break
    if os.path.isfile(os.path.join(dir_axion, path)):
        train_data[count, 0, :, :] = np.load(os.path.join(dir_axion, path), allow_pickle=True)[0]
        count += 1
for path in os.listdir(dir_cdm):
    if count==val_size *2:
        break
    if os.path.isfile(os.path.join(dir_cdm, path)):
        train_data[count, 0, :, :] = np.load(os.path.join(dir_cdm, path), allow_pickle=True)
        count += 1
for path in os.listdir(dir_no):
    if count==val_size*3:
        break
    if os.path.isfile(os.path.join(dir_no, path)):
        train_data[count, 0, :, :] = np.load(os.path.join(dir_no, path), allow_pickle=True)
        count += 1
print(train_data.shape)
```
